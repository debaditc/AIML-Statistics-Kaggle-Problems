## Analysis of Violin graph on Product and Fitness

```python
n_cardio_data_df = cardio_data_df.copy()

# For violin graph , we need to have numerical value for Product
# First , we convert the Product into numbers and put into a new dervied field "Product_number"

# Create conditions in list for Products
conditions = [
    (n_cardio_data_df['Product'] == 'TM195'),
    (n_cardio_data_df['Product'] == 'TM498'),
    (n_cardio_data_df['Product'] == 'TM798')
    ]

# Assign values 1,2,3 for each Product
# 1 -> TM195
# 2 -> TM498
# 3 -> TM798
values = [1,2,3]

# Create new column (Product_number) and assign values "1,2,3" accordingly
n_cardio_data_df['Product_number'] = np.select(conditions, values)

# Use the modifed dataframe (with derived column) to create violin plot (Product vs Fitness)
sns.violinplot(x=n_cardio_data_df.Product_number, y=cardio_data_df['Fitness'])
```

**Observation from violin graph on Product and Fitness**
- People who rated their fitness  as 4 and above are inclined towards TM798
- For general , wide usage and starters - TM195 is preferred

## Create groups on various products based on age and gender

```python
groups = cardio_data_df.groupby(cardio_data_df.Product)
TM195_df = groups.get_group('TM195')
TM498_df = groups.get_group('TM498')
TM798_df = groups.get_group('TM798')
```

```python
plt.hist(TM195_df.Age)
plt.hist(TM498_df.Age)
plt.hist(TM798_df.Age)
plt.legend(['TM195','TM498','TM798']);
plt.show();

plt.hist(TM195_df.Gender)
plt.hist(TM498_df.Gender)
plt.hist(TM798_df.Gender)
plt.legend(['TM195','TM498','TM798']);
plt.show();


plt.hist(TM195_df.Income)
plt.hist(TM498_df.Income)
plt.hist(TM798_df.Income)
plt.legend(['TM195','TM498','TM798']);
plt.show();


plt.hist(TM195_df.MaritalStatus)
plt.hist(TM498_df.MaritalStatus)
plt.hist(TM798_df.MaritalStatus)
plt.legend(['TM195','TM498','TM798']);
plt.show();
```

**Observation on age and products**
- TM195 is preferred in age group between 20 and 30 and widely spreads to 50
- TM498 is more preferred in age group between 20 and 45. But not as widely spread as other 2 options
- TM798 is concentrated over 15-30 age group but widely spreads as TM195

**Observation on gender and products**
- TM798 is highly perferred by Male and not using TM498
- Incase of females, TM498 is mostly preferred


**Observation on income and products**
- TM195 and TM498 are more popular with income range between 30000 to 50000
- TM798 seems more expensive but more popular with people having higher income 


**Observation on Marital Status and products**
- Partnered are working out more than single customer
- Both in partnered and single - TM798 is more popular as compared with TM195 and TM498

## Cat Plot on Fitness-Product and Fitness-Gender

```python
sns.catplot(x="Fitness", col="Gender", col_wrap=4,
                data=cardio_data_df,
                kind="count")

sns.catplot(x="Fitness", col="Product", col_wrap=4,
                data=cardio_data_df,
                kind="count")
```

**Observation**

- 2nd chart : Fitness level with 3 (moderate and regular use) , customer prefer more on TM195.
